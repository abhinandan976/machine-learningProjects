```python
#Encoding the categorical feature by label encoder
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

categorical_featuers_encoded = categorical_features.copy()
for col in categorical_featuers_encoded:
 categorical_featuers_encoded[col] = label_encoder.fit_transform(categorical_featuers_encoded[col])
```

```python
print("Enter the cuisine for your preference")
cuisine_preference = input().strip('"')
print("Enter the maxiumum price for the user")
max_price = int(input())
#filtering the data into user choices
if 'cuisine' in data.columns:
  recommended_restaurants = data[
    (data['cuisine'] == cuisine_preference)&(data['price']<=max_price)
  ]
  print(recommended_restaurants)
else:
  print("Column 'cuisine' not found in the data")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
feature_matrix = categorical_featuers_encoded.values
similarity_matrix = cosine_similarity(feature_matrix)

preferred_restaurant_index = 0
similar_restaurants_indices = similarity_matrix[preferred_restaurant_index].argsort()[::-1][1:]
print("Top 5 similar restuarants:")
for index in similar_restaurants_indices[:5]:
  print(data.iloc[index])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

relevant_columns = ['Restaurant Name', 'City', 'Cuisines', 'Price range', 'Aggregate rating']

# Extract relevant features
features = data[relevant_columns].copy()

features['combined_features'] = features.apply(lambda row: ' '.join(row.dropna().astype(str)), axis=1)

# Create a TF-IDF Vectorizer to convert text data into numerical data
tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(features['combined_features'])

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Function to get recommendations based on user preferences
def get_recommendations(preferences, cosine_sim=cosine_sim):
    user_tfidf = tfidf.transform([preferences])
    user_similarities = linear_kernel(user_tfidf, tfidf_matrix)
    sim_scores = list(enumerate(user_similarities[0]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    top_indices = [i[0] for i in sim_scores[:10]]
    return data.iloc[top_indices][['Restaurant Name', 'City', 'Cuisines', 'Price range', 'Aggregate rating']]

sample_preferences = "Japanese Makati City 3 4.5"
recommended_restaurants = get_recommendations(sample_preferences)
print(recommended_restaurants)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
relevant_columns = ['Restaurant Name', 'City', 'Cuisines','Price range', 'Aggregate rating']
features = data[relevant_columns].copy()

# Convert 'Price range' and 'Aggregate rating' to strings before concatenation and fill NaN with empty strings
features['combined_features'] = features['Restaurant Name'] + '  '+ features['City'] + '  ' + features['Cuisines'] + '  ' + features['Price range'].astype(str) + '  ' + features['Aggregate rating'].astype(str)
features['combined_features'] = features['combined_features'].fillna('') # Fill NaN values with empty strings

tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(features['combined_features'])

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

def get_recommendations(preferences, cosine_sim=cosine_sim):
  user_tfidf = tfidf.transform([preferences])
  user_similarities = linear_kernel(user_tfidf, tfidf_matrix)

  sim_scores = list(enumerate(user_similarities[0]))
  sim_scores = sorted(sim_scores, key =lambda x:x[1],reverse=True)

  top_indices = [i[0] for i in sim_scores[:10]]

  return data.iloc[top_indices][['Restaurant Name', 'City','Cuisines','Price range','Aggregate rating']]

sample_preferences = "Japanese Makati City 3 4.5"
recommended_restaurants = get_recommendations(sample_preferences)
print(recommended_restaurants)
```
